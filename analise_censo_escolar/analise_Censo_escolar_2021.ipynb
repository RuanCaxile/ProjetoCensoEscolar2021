```python
# importando a biblioteca Pandas
import pandas as pd
```

```python
# lendo o arquivo .csv do Censo Escolar 2021 do INEP, e a partir disso chamando o Data Frame geral com todas as informações presentes no arquivo
microDadosCenso = pd.read_csv('C:\\Users\\ruanc\\OneDrive\\Área de Trabalho\\Portfolio\\ProjetoCensoEscolar2021\\microdados_censo_escolar_2021\\microdados_ed_basica_inep\\dados\\microdados_ed_basica_2021.csv', sep=';', encoding='latin1')
microDadosCenso
```

```python
# observando todos os nomes das colunas presentes no Data Frame microDadosCenso.
microDadosCenso.columns.values
```

```python
# irei trabalhar com essas colunas da base de dados do censo
colunasDeInteresseDoProjeto = [ 
    'NO_REGIAO', 'NO_UF','NO_MUNICIPIO','QT_MAT_BAS', 'TP_DEPENDENCIA', 'TP_LOCALIZACAO',
    'TP_SITUACAO_FUNCIONAMENTO','IN_MEDIACAO_PRESENCIAL',
    'IN_MEDIACAO_SEMIPRESENCIAL', 'IN_MEDIACAO_EAD', 'IN_EQUIP_NENHUM', 'IN_DESKTOP_ALUNO',
    'QT_DESKTOP_ALUNO', 'IN_COMP_PORTATIL_ALUNO', 'QT_COMP_PORTATIL_ALUNO',
    'IN_TABLET_ALUNO', 'QT_TABLET_ALUNO', 'IN_INTERNET',
    'IN_INTERNET_APRENDIZAGEM', 'IN_ACES_INTERNET_DISP_PESSOAIS'
]
```

```python
# realizando o filtro do Data Frame microDadosCenso para ter um Data Frame com somente as colunas que são do interesse do Projeto.
dataFrameInicialComColunasDeInteresse = microDadosCenso.filter(colunasDeInteresseDoProjeto)
dataFrameInicialComColunasDeInteresse
```

```python
# apenas verificando se no conjunto de todas as escolas que compõe o Censo alguma possuía informações NaN em alguma das colunas de interesse 
verificandoColunasDeInteresseVazia = dataFrameInicialComColunasDeInteresse.dropna()
verificandoColunasDeInteresseVazia
```

## Agora, nosso projeto realmente começa!
Agora que já temos segurança no escopo do projeto, vamos selecionar o Data Frame que iremos trabalhar de fato para gerar o arquivo 'AnaliseDigitalCensoEscolar2021.csv'.

```python
# definindo o Data Frame do Projeto, que será somente os que possuem situação de funcionamento ativa.
EscolasComSituacaoDeFuncionamentoAtivas = dataFrameInicialComColunasDeInteresse[dataFrameInicialComColunasDeInteresse['TP_SITUACAO_FUNCIONAMENTO']==1]
EscolasComSituacaoDeFuncionamentoAtivas
```

```python
# modificando os valores numéricos da categoria das escolas para valores categóricos!
EscolasComSituacaoDeFuncionamentoAtivas['TP_DEPENDENCIA']=EscolasComSituacaoDeFuncionamentoAtivas['TP_DEPENDENCIA'].replace({1:'Federal', 2:'Estadual', 3:'Municipal', 4:'Privada'})
```

```python
# modificando os valores numéricos da localização das escolas para valores categóricos!
EscolasComSituacaoDeFuncionamentoAtivas['TP_LOCALIZACAO'] = EscolasComSituacaoDeFuncionamentoAtivas['TP_LOCALIZACAO'].replace({1:'Urbana', 2:'Rural'})
```

```python
# como já sabemos que no Data Frame EscolasComSituacaoDeFuncionamentoAtivas são todas com a coluna TP_SITUACAO_FUNCIONAMENTO com valores
# iguais a 1, não precisamos dessa coluna.
EscolasComSituacaoDeFuncionamentoAtivas.drop('TP_SITUACAO_FUNCIONAMENTO',axis=1, inplace=True)
```

```python
# verificando a quantidade de escolas ativas que ainda possuiam mediação EAD no ano de 2021.
EscolasComSituacaoDeFuncionamentoAtivas['IN_MEDIACAO_EAD'].value_counts()
```

```python
listaComColunasAsofrerReplace = [
    'IN_MEDIACAO_PRESENCIAL','IN_MEDIACAO_SEMIPRESENCIAL', 'IN_MEDIACAO_EAD', 'IN_EQUIP_NENHUM', 'IN_DESKTOP_ALUNO',
    'IN_TABLET_ALUNO', 'IN_COMP_PORTATIL_ALUNO', 'IN_INTERNET','IN_INTERNET_APRENDIZAGEM', 'IN_ACES_INTERNET_DISP_PESSOAIS'
]

for coluna in listaComColunasAsofrerReplace:
    EscolasComSituacaoDeFuncionamentoAtivas[coluna] = EscolasComSituacaoDeFuncionamentoAtivas[coluna].replace({1: 'Sim', 0:'Não'})
```

```python
# finalmente, basta transformar o Data Frame principal, que já teve seus dados tratados e devidamente categorizados, em um arquivo .csv
EscolasComSituacaoDeFuncionamentoAtivas.to_csv(r'C:\\Users\\ruanc\\OneDrive\\Área de Trabalho\\Portfolio\\ProjetoCensoEscolar2021\\microdados_censo_escolar_2021\\microdados_ed_basica_inep\\dados\\AnaliseDigitalCensoEscolar2021.csv', index=False)
```
